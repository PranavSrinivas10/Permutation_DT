```python
class node:
    def __init__(self,attr):
        self.children=[]
        self.attr=attr
        self.purity=[0,0]
# defining custom node class to print decision tree
```

```python
def checkhomogenous(arr):
    n = len(arr)
    if n == 0:
        return True
    cur = arr[0]
    for num in arr:
        if num != cur:
            return False
    return True
```

```python
def nsrps(arr, k):
    if checkhomogenous(arr):
        return 0
        
    n = len(arr)
    freq = {}
    mxnum = max(arr)
    mxfreq = 0
    mxtup = ()
    for i in range(n - k + 1):
        cur = tuple(arr[i:i+k])
        if cur in freq:
            freq[cur] += 1
        else:
            freq[cur] = 1
        if(freq[cur] > mxfreq):
            mxfreq = freq[cur]
            mxtup = cur
    
    narr = []
    i = 0
    while (i < n):
        cur = tuple(arr[i:i+k])
        if(cur == mxtup):
            narr.append(mxnum + 1)
            i = i + k
        else:
            narr.append(arr[i])
            i += 1
    return 1 + nsrps(narr, k)
```

```python
def cnt(df,index):
    di={}
    for i in df[index]:
        if i not in di:
            di[i]=1
        else:
            di[i]+=1
    return di
# function to count number of instances of each unique value for an attribute/feature(df[index])
# index refers to the feature number (like f1 or f2)
```

```python
def feature(df,index):
    di=cnt(df,index)
    feature_entropy=0
    for i in di:
        l=[]
        for j in range(df.shape[0]):
            if(df[index][j]==i):
                l.append(df[df.shape[1]-1][j])
        feature_entropy+=(di[i]/df.shape[0])*nsrps(l,2)
    return feature_entropy
# calculates ETC(S|A) for each attribute/feature A in the dataset
```

```python
def split(df):
    l=[x for x in df[df.shape[1]-1]]
    hs=nsrps(l,2)
    feature_size=df.shape[1]-1
    mgain=-1
    mfeature=0
    for i in range(feature_size):
        gain=hs-feature(df,i)
        if(gain>mgain):
            mgain=gain
            mfeature=i
    return mfeature
# chooses best feature to split on at a particular level in the decision tree based on Information gain value.
```

```python
def printTree(node, level=0):
    if node != None:
        x=node.children[::-1]
        flag=True
        for i in range(len(x)):
            if(i==len(x)//2):
                print(' ' * 15 * level + '-> ',node.attr,node.purity)
                flag=False
            printTree(x[i], level + 1)
        if(flag):
            print(' ' * 15 * level + '-> ',node.attr,node.purity)
# travesrsing the constructed tree to print the nodes in the correct order.
# node at the leftmost indentation is root node and nodes at the rightmost indentation are leaf nodes
```
